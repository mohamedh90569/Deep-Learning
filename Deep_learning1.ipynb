```python
df=df.dropna()
```

```python
df=pd.get_dummies(df,columns=['Origin'],drop_first=True)
df
```

```python
x=df.drop('Horsepower',axis=1)
y=df['Horsepower']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
Scaler=StandardScaler()
Scaler.fit(x_train)
x_train=Scaler.transform(x_train)
x_test=Scaler.transform(x_test)
```

# **Start with Simple Model**

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau,ModelCheckpoint
```

```python
x_train.shape
```

```python
x_train.shape[1]
```

```python
model = Sequential()
model.add(Dense(units= 100 , activation ='relu',input_shape=[x_train.shape[1]]))
model.add(Dense(units=1))
```

```python
model.compile(loss='mean_squared_error',optimizer=Adam())
```

```python
Ird=ReduceLROnPlateau(monitor='val_loss', patience=200 ,verbose=1 , factor=0.75, min_lr=1e-4)
```

```python
mcp=ModelCheckpoint('model.h5')
es= EarlyStopping(verbose=1,patience=600)
```

```python
model.summary()
```

```python
x=model.fit(x_train,y_train,validation_split=0.2,batch_size=32, epochs=1000,callbacks=[Ird,mcp,es])
```

```python
model.evaluate(x_test,y_test)
```

```python
y_pred=model.predict(x_test)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```
